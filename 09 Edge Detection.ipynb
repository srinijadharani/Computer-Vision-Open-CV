## Edge Detection: Canny Edge Detector Algorithm

Steps:

    1. Gaussian filter to smooth out the edges and remove the noise
    2. Find intensity gradients
    3. Non-maximum suppression to get rid of spurious (false) response to edge detection
    4. Double threshold to determine potential edges
    5. Track edge by hysteresis

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread("../DATA/sammy_face.jpg")
plt.imshow(img)
```

```python
edges=cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges=cv2.Canny(image=img, threshold1=0, threshold2=255)
plt.imshow(edges)
```

```python
med_value=np.median(img)
med_value
```

```python
lower=int(max(0, 0.7*med_value))
upper=int(min(255, 1.3*med_value))
```

```python
edges=cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
upper
```

```python
edges=cv2.Canny(image=img, threshold1=lower, threshold2=upper+150)
plt.imshow(edges)
```

```python
blurred_img=cv2.blur(src=img, ksize=(5,5))
edges=cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper+70)
plt.imshow(edges)
```
